# Using ocr_helper Package

```python
import matplotlib.pyplot as plt
from ocr_helper import get_digit_crops
```

```python
digit_crops = get_digit_crops('pic/53.png')
print(f"Number of Digits Cropped: {len(digit_crops)}")
```

```python
for i, digit in enumerate(digit_crops):
    plt.figure()
    plt.imshow(digit, cmap='gray')
```
